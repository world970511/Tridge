```python
import pandas as pd
import numpy as np
import re
import plotly.express as px
```

```python
# 기입된 Product 종류 및 개수

product_list=list(set(data['Product'].values.tolist())) 
product_list.sort()
print('종류:',product_list,'\n개수:',len(product_list))
```

## Product-Country 그래프

```python
#기입된 국가
country_list=list(set(data['Country'].values.tolist())) 
country_list.sort()
print('종류:',country_list,'\n개수:',len(country_list))
```

```python
#신규 데이터프레임
df= pd.DataFrame(columns=['product','country','count'])

for product in product_list:
    d=data.loc[data["Product"]==product]
    for country in country_list:
        count=len(d[d['Country']== country])
        if count>0:
            df.loc[len(df)]=[product,country,count]
#시각화 
fig = px.bar(df, x="product", y="count",
             hover_data=['country'],color="country"
            ,title="Product-Country")
fig.show()
```

## Arrange

##### 제거/ 변경해야 하는 것
- ['Grades']: 'Chestnut / Variety / Unspecified' or 'No Grade' or Null -> Unverifiable ( 잘못 기입되었거나 등급이 없는 것은 확인 불가로 변경)

- ['Grades']: 'Avocado / Grade / Third Quality - New' ->'Avocado / Grade / Third Quality'(등급과 관련 없는 정보 제거)

- ['Grades']: 'Strawberry / Grade / For Juice', 'Strawberry / Grade / For Juice/ Strawberry / Grade / Industrial' ->'Industrial' 로 변경

- ['Variety']: Spring Garlic (= Green Garlic=Young Garlic) 같은 동일한 종인데 다른 이름으로 처리되어 있는것 변경

- ['Region']: 주/ 도시 명 외에 제거

### 1차 정리

```python
test_data=data

#Variety/Grades 1차 정리
test_data['Variety']=test_data['Variety'].apply(lambda x: x.split('/ ')[-1] if '/' in x else x)
test_data['Grades']=test_data['Grades'].apply(lambda x:''.join( x.split('/ ')[-1]) if '/ ' in x else x)

test_data.head()
```

### 2차 정리

```python
"""데이터 2차 정리
- 'Grades': 특수문자(. 제외) + 문자 조합 제거 (- New)/Unverifiable -> No Grade 변경
- 'Region': 주/ 도시 명 외에 제거.
"""
import re

#특수문자(. 제외) + 문자 조합 제거 (- New)/Unverifiable -> No Grade 변경
change=['No Grade','Unspecified','For Juice']
test_data['Grades']=test_data['Grades'].apply(lambda x: 'Unverifiable' if x in change else x)
test_data['Grades']=test_data['Grades'].apply(lambda x: re.sub('[-!#$&*].+',"",x) )
test_data['Grades']=test_data['Grades'].apply(lambda x: x[:-1] if x[-1]==' ' else x)
```

```python
#Spring Garlic (= Green Garlic=Young Garlic) 같은 동일한 종인데 다른 이름으로 처리되어 있는것 변경
dic={'Green Garlic':'Spring Garlic','Young Garlic':'Spring Garlic','Spring Garlic':'Spring Garlic',
     'Cerises':'Cherry Varieties','Cherry Varieties':'Cherry Varieties',
     'Kamaroza':'Camarosa','Camarosa':'Camarosa'}
def fix_variety(x):
  dic={'Green Garlic':'Spring Garlic','Young Garlic':'Spring Garlic','Spring Garlic':'Spring Garlic',
     'Cerises':'Cherry Varieties','Cherry Varieties':'Cherry Varieties',
     'Kamaroza':'Camarosa','Camarosa':'Camarosa'}
  for i in dic.keys():
       if i in x:
         x=dic[i]
  return x
  
test_data['Variety']=test_data['Variety'].apply(fix_variety)
```

```python
#'Region': 주/ 도시 명 외에 제거.

def fix_region(x):
    stopword=['Province']
    li=x.split(' ')

    for word in stopword:
      if word in li:
        li.remove(word)
    return ' '.join(li)
  

test_data['Region']=test_data['Region'].apply(fix_region)
```
